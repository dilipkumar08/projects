```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae
```

```python
df=pd.read_csv('bigmart_data.csv')
```

```python
df.isnull().sum()
```

```python
df=df.sample(frac=1,ignore_index=True)
```

```python
df
```

```python
div=df.shape[0]//4
```

```python
train=df.loc[:3*div+1,:]
test=df.loc[3*div+1:]
```

### Simple Mean

```python
test['simple_mean']=train['Item_Outlet_Sales'].mean()
```

```python
print(mae(test['Item_Outlet_Sales'],test['simple_mean']))
```

```python
out_typ=pd.pivot_table(train,values='Item_Outlet_Sales',
                      index=['Outlet_Type'],aggfunc='mean')
out_typ
```

```python
test['Out_type_mean']=0
for i in train['Outlet_Type'].unique():
    test['Out_type_mean'][test['Outlet_Type']==str(i)]=train['Item_Outlet_Sales'][train['Outlet_Type']==str(i)].mean()
```

```python
mae(test['Item_Outlet_Sales'],test['Out_type_mean'])
```

```python
pd.pivot_table(train,values='Item_Outlet_Sales',index='Outlet_Establishment_Year',aggfunc='mean')
```

```python
test['Outlet_Type_Year_Mean']=0
```

```python
test['Outlet_Type'][(test['Outlet_Type']=='Grocery Store') & (test['Outlet_Establishment_Year']=='2009')]
```

```python
train['Outlet_Location_Type']=train['Outlet_Location_Type']
train.Outlet_Location_Type.unique()
```

```python
for year in train.Outlet_Establishment_Year.unique():
    for type in train.Outlet_Type.unique():
        for location in train.Outlet_Location_Type.unique():
            test['Outlet_Type_Year_Mean'][(test['Outlet_Type']==str(type)) & (test['Outlet_Location_Type']==location) & (test['Outlet_Establishment_Year']==int(year))]=train['Item_Outlet_Sales'][(train['Outlet_Type']==str(type)) & (train['Outlet_Establishment_Year']==int(year)) & (train['Outlet_Location_Type']==location)].mean()
```
