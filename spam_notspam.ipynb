```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import re
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from nltk.stem import PorterStemmer
np.random.seed(42)
```

```python
data=pd.read_csv("email_classification.csv")
```

```python
data
```

```python
for i in range(179):
    print(data["email"][i], data["label"][i])
```

```python
def preprocess_text(text):
    text=text.lower()
    text=re.sub(r'\W+',' ',text)
    text=re.sub(r'\d+','',text)
    text = text.split()
    text=list(set(text))
    return text

```

```python
data["email_1"]=data["email"].apply(preprocess_text)
```

```python
word_collection=[]
for i in data["email_1"]:
    for j in i:
        print(j)
        word_collection.append(j)
```

```python
data["label_1"]=data["label"].map({'ham':0,'spam':1})
```

```python
data
```

```python
def bag_of_words(text,word_dict):
    bag=np.zeros(len(word_dict))
    for i in text:
        if i in word_dict:
            bag[word_dict[i]]+=1
    return bag
```

```python
X=np.array([bag_of_words(email,word_dict) for email in data["email_1"]])
Y=np.array(data["label_1"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.7,shuffle=True)
```

```python
model=SVC(kernel="linear")
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
accuracy_score(y_test,model.predict(x_test))
```

```python
precision_score(y_test,model.predict(x_test))
```

```python
confusion_matrix(y_test,model.predict(x_test),labels=[1,0])
```

```python
pd.crosstab(y_test,model.predict(x_test))
```

```python
def stemming(text):
    ps=PorterStemmer()
    for t in range(len(text)):
        text[t]=ps.stem(text[t])
    return text
        
```
